#### Calculate the sample mean and variance of the variables: price, rooms, surface area and bathrooms (★★★)

#### Normalize the variable of prices for each population and plot the 4 histograms in the same plot (★★★) 

```python
import pandas as pd
ds = pd.read_csv('assets/real_estate.csv', sep=';')
### Normalize prices
def normalize(series):
    return (series - series.min())/(series.max() - series.min())

ds['normalized_price'] = ds.groupby('level5')['price'].transform(normalize)
ds

### Histograms
locations = ['Fenlabrada', 'leganés', 'Getafe', 'Alcorcón']

for location in locations:
    ds[ds['level5'] == location]['normalized_price'].hist(alpha=0.5, label=location)

plt.legend()
plt.show()
```

#### What can you say about the price per square meter between the towns of Getafe and Alcorcón?

```python
import pandas as pd
ds = pd.read_csv('assets/real_estate.csv', sep=';')
pd.set_option('display.max_columns', None)

# Getafe
mean_price_getafe = ds[ds['level5'] == 'Getafe']['price'].mean()
mean_sqrMt_Getafe = ds[ds['level5'] == "Getafe"]['surface'].mean()
mean_price_per_sqrMt_Getafe = mean_price_getafe / mean_sqrMt_Getafe

print("The mean price per square meter in Getafe is:",mean_price_per_sqrMt_Getafe)

# Alcorcón
mean_price_Alc = ds[ds['level5'] == 'Alcorcón']['price'].mean()
mean_sqrMt_Alc = ds[ds['level5'] == "Alcorcón"]['surface'].mean()
mean_price_per_sqrMt_Alc = mean_price_Alc/ mean_sqrMt_Alc

print("The mean price per square meter in Alcorcón is:",mean_price_per_sqrMt_Alc)

if mean_price_per_sqrMt_Getafe < mean_price_per_sqrMt_Alc:
    print("Getafe is more affordable than Alcorcón")
elif mean_price_per_sqrMt_Getafe > mean_price_per_sqrMt_Alc:
    print('Alcorcón is more affordable than Getafe')
else:
    print('Both cities have the same affordability')
```

#### Reload the data, is the average of "Valdemorillo" and "Galapagar" prices the same?

```python
import pandas as pd
ds = pd.read_csv('assets/real_estate.csv', sep=';')
pd.set_option('display.max_columns', None)

avg_price_Vald = ds[ds['level5'] == "Valdemorillo"]['price'].mean()
avg_price_Galapagar = ds[ds['level5'] == "Galapagar"]['price'].mean()

if avg_price_Vald == avg_price_Galapagar:
    print('The average of prices in Valdemorillo and Galapagar are the same')
else:
    print('The average of prices in Valdemorillo and Galapagar are NOT the same')
    if avg_price_Vald > avg_price_Galapagar:
        percentage = ((100 * avg_price_Vald) / avg_price_Galapagar) - 100
        print ('The average price of Valdemorillo is',round(percentage),"percent higher than Galapagar's")
    else:
        percentage = (100 * avg_price_Vald) / avg_price_Galapagar
        print ('The average price of Valdemorillo is',round(percentage),"percent lower than Galapagar's")
```

#### Could you say there is a relation between surface and price? (★★☆)

```python
import matplotlib.pyplot as plt
import pandas as pd
ds = pd.read_csv('assets/real_estate.csv', sep=';')

#Ininitally noted that some properties had values way off on x axis, getting rid of them temporarily for better visualziation
threshold = 1000
mask = ds['surface'] < threshold
filtered_ds = ds[mask]

plt.scatter(filtered_ds['surface'], filtered_ds['price'], alpha=0.25)
plt.xlabel('Surface')
plt.ylabel('Price')
plt.show()

```
